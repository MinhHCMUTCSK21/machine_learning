```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('matches.csv', index_col=0)
```

```python
df['team'].value_counts()
```

```python
df[df['team']=='Liverpool']
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['venue_code'] = df['venue'].astype('category').cat.codes
df
```

```python
df['opp_coed'] = df['opponent'].astype('category').cat.codes
df
```

```python
df['hour'] = df['time'].str.split(':').str[0].astype(int)
df
```

```python
df['day_code'] = df['date'].dt.dayofweek
```

```python
df['target'] = (df['result'] == 'W').astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = df[df['date'] < '2022-01-01']
test = df[df['date'] >= '2022-01-01']

predictors = ['venue_code', 'opp_coed', 'hour', 'day_code']

rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])

preds
```

```python
from sklearn.metrics import accuracy_score

acc= accuracy_score(test['target'], preds)

acc
```

```python
combined = pd.DataFrame(dict(actual=test['target'], prediction=preds))
```

```python
pd.crosstab(index = combined['actual'], columns= combined['prediction'])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test['target'], preds)
```

```python
grouped_matches = df.groupby('team')
```

```python
group = grouped_matches.get_group('Liverpool')
group
```

```python
def rolling_avg(group, col, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[col].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group    
```

```python
cols = ['gf','ga','sh' ,'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f"{col}_rolling" for col in cols]
```

```python
new_cols
```

```python
rolling_avg(group, cols, new_cols)
```

```python
matches_rolling = df.groupby('team').apply(lambda x : rolling_avg(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
def make_prediction (df, predictors): 
    train = df[df['date'] < '2022-01-01']
    test = df[df['date'] >= '2022-01-01']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], prediction=preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

```python
combined, precision = make_prediction(matches_rolling, predictors + new_cols)
```
